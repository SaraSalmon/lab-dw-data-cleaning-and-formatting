## Exercise 1: Cleaning Column Names

To ensure consistency and ease of use, standardize the column names of the dataframe. Start by taking a first look at the dataframe and identifying any column names that need to be modified. Use appropriate naming conventions and make sure that column names are descriptive and informative.

```python
import numpy as np
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv")
df
```

## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".

```python
insurance_2_df["gender"].unique()
```

```python
insurance_3_df = insurance_2_df.copy()
gender_mapping = {'Male': 'M', 
                  'female': 'F',
                   'nan': 'nan',
                  'F':'F',
                  'M': 'M',
                  'Femal':'F'
                  }

insurance_3_df['gender'] = insurance_3_df['gender'].map(gender_mapping)
insurance_3_df.head()
```

```python
insurance_3_df["gender"].unique()
```

State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"

```python
insurance_4_df = insurance_3_df.copy()
insurance_4_df["state"].unique()
```

```python
gender_mapping = {'Washington': 'Washington', 
                  'Arizona': 'Arizona',
                  'Nevada': 'Nevada',
                  'California':'California',
                  'Oregon': 'Oregon',
                  'Cali':'California',
                  'AZ':'Arizona',
                  'WA':'Washington',
                  'nan':'nan'
                  }

insurance_4_df['state'] = insurance_4_df['state'].map(gender_mapping)
insurance_4_df.head()
```

```python
insurance_4_df["state"].unique()
```

```python
insurance_4_df.shape
```

In education, "Bachelors" could be replaced by "Bachelor"

```python
insurance_5_df = insurance_4_df.copy()
insurance_5_df["education"].unique()
```

```python
gender_mapping = {'Master': 'Master', 
                  'Bachelor': 'Bachelor',
                  'High School or Below': 'High School or Below',
                  'College':'College',
                  'Bachelors': 'Bachelor',
                  'Doctor':'Doctor',
                  'nan':'nan'
                  }

insurance_5_df['education'] = insurance_5_df['education'].map(gender_mapping)
insurance_5_df.head()
```

```python
insurance_5_df["education"].unique()
```

```python
insurance_5_df.shape
```

In Customer Lifetime Value, delete the % character

```python
insurance_6_df = insurance_5_df.copy()
insurance_6_df['customer_lifetime_value'] = insurance_6_df['customer_lifetime_value'].str.replace('%', '', regex=False)
insurance_6_df
```

In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"

```python
insurance_7_df = insurance_6_df.copy()
insurance_7_df["vehicle_class"].unique()
```

```python
gender_mapping = {'Four-Door Car': 'Four-Door Car', 
                  'Two-Door Car': 'Two-Door Car',
                  'SUV': 'SUV',
                  'Luxury SUV':'Luxury',
                  'Sports Car': 'Luxury',
                  'Luxury Car':'Luxury',
                  'nan':'nan'
                  }

insurance_7_df['vehicle_class'] = insurance_7_df['vehicle_class'].map(gender_mapping)
insurance_7_df.head()
```

```python
insurance_7_df["vehicle_class"].unique()
```

```python
insurance_7_df.shape
```

```python
# Como es un valor parecido, se aplica el forward fill: 
insurance_10_df = insurance_9_df.copy()
insurance_10_df ["gender"] = insurance_10_df["gender"].fillna(method='ffill')
insurance_10_df["gender"].isnull().sum()
```

```python
insurance_10_df ["gender"] = insurance_10_df["gender"].fillna(method='bfill')
insurance_10_df
```

```python
# Para los customer lifetime value, se rellenarÃ¡n con la media de todos:
insurance_11_df = insurance_10_df.copy()
media = insurance_11_df["customer_lifetime_value"].mean()
insurance_11_df["customer_lifetime_value"] = insurance_11_df["customer_lifetime_value"].fillna(media) # Relleno los Nnas de Age con la media 
```
